# Q1. Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop ?

 ANSWER

1. Publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop because the real value of data comes from collaboration, accessibility, and decision-making, not just visualization.

2. Dashboards in Desktop are limited to the creator

* When you build reports or dashboards in Microsoft Power BI Desktop, they stay on your local machine. That means: Only you can view and interact with them

3.  Publishing enables sharing across teams
* When you publish your report to Power BI Service:
* Reports become accessible via browser or mobile
* Teams across departments can view the same data

4. Better security and access control
* In Power BI Service, you can: Control who sees what data
* Set permissions (viewer, contributor, admin)

# Q4. List and explain the steps involved in publishing a report from Power BI Desktop to Power BI Service in the correct sequence

ANSWER

1. Prepare the Report

* Create and finalize your report in Power BI Desktop (add visuals, clean data, check accuracy).

2. Save the Report

* Save the file in .pbix format on your computer.

3. Sign In to Power BI

* Open Power BI Desktop and sign in using your Microsoft account.

4. Click on Publish

* Go to the Home tab and click the “Publish” button.

5. Select Destination Workspace

* Choose where you want to publish (e.g., My Workspace or a team workspace).

6. Confirm Publishing

* Click Select/OK to start uploading the report to Power BI Service.

7. Wait for Upload Completion

* A confirmation message appears once the report is successfully published.

8. Open in Power BI Service

* Click the link provided or go to Power BI Service in your browser to view the report online.

# Q6. Explain the difference between the following three sharing methods in Power BI Service:
 * Direct Report Sharing
* Workspace Access
* Power BI Apps
* Also mention one use case for each.

ANSWER

1. Direct Report Sharing

* Explanation: Sharing a specific report directly with individual users via email or link.

* Access Level: Limited to selected users only.

* Control: Basic permissions (view or edit).

* Use Case: Sharing a sales report with a specific manager for quick review.


2. Workspace Access

* Explanation: Providing access to a workspace where multiple reports, dashboards, and datasets are stored.

* Access Level: Multiple users (team-based access).

* Control: Role-based permissions (Admin, Member, Contributor, Viewer).

* Use Case: A project team collaborating on multiple reports in the same workspace.


3. Power BI Apps

* Explanation: Packaging and distributing reports and dashboards as an app to a large audience.
* Access Level: Broad audience (entire organization or large groups).

* Control: Centralized and controlled distribution.

* Use Case: Publishing a company-wide performance dashboard for all employees.

# Q7. Describe the four workspace roles available in Power BI and explain how access control is managed using these roles

ANSWER

1. Admin

* Description: Full control over the workspace

* Permissions: Add/remove users, Assign roles , Edit/delete all content

* Access Level: Highest
* Use Case: Team leader or project owner managing everything

2. Member

* Description: Can collaborate and manage content

* Permissions: Create, edit, and delete reports/dashboards , Share content

* Access Level: High (but cannot manage all settings like Admin)

* Use Case: Core team members working on reports.


3. Contributor

* Description: Can create and modify content but with limited sharing

* Permissions: Add and edit reports, Cannot share or manage user access fully

* Access Level: Medium

* Use Case: Report developers or analysts



4.  Viewer

* Description: Read-only access

* Permissions: View and interact with reports , Cannot edit or share

* Access Level: Lowest

* Use Case: End users or stakeholders

5. How Access Control is Managed

* Access is controlled by assigning these roles to users in a workspace

* Each role defines what actions a user can or cannot perform

* Organizations ensure:

Security → Only authorized users can edit/share

Collaboration → Teams work together efficiently

Data integrity → Limited editing rights prevent errors

# Q8. Why are Power BI Apps considered more secure and scalable than direct sharing? Explain this from a governance and enterprise usage perspective

ANSWER

1. Power BI Apps are generally considered more secure and scalable than direct sharing because they align better with enterprise governance, controlled access, and structured distribution of content.

2. Better Control Over Content

* Power BI Apps are published from a workspace and only a few authorized users can edit or update them. End users can only view the content.

 Ensures reports are not accidentally modified

 Maintains a single “official version” of reports

* In direct sharing, multiple users may have edit access, which increases the risk of errors and data misuse.


3. Scalable Access Management

* Apps allow access to be given to large groups (like departments or teams).

 Works well for large organizations

 Reduces manual effort

* Direct sharing requires adding users one by one, which is not practical at scale.



4. Better Compliance & Auditing

* From a governance perspective, Apps make it easier to:

Track who has access

Monitor usage

Enforce organizational policies

* Direct sharing can lead to:

“Permission sprawl”

Difficulty in auditing who has access to what

# Q9. Explain Row-Level Security (RLS) in Power BI.How does RLS ensure that different users see different data using the same report?

ANSWER

1. Row-Level Security (RLS) is a feature that restricts data access for different users. It ensures that users only see the data they are allowed to see, even when using the same report.

2.  HOW WORK

3. Create Roles

* Define roles like Sales Manager, HR, Region West, etc.

4. Apply Filters to Roles

* Add rules (filters) on data tables

* Example: West Manager → sees only West region data

* HR → sees only employee data

5. Assign Users to Roles

* Map users (or groups) to specific roles

6. Automatic Data Filtering

* When a user opens the report, Power BI checks their role

* Data is filtered automatically based on assigned rules



7. How Different Users See Different Data

* Same report is shared with everyone

* RLS applies filters in the background

* Each user sees only relevant rows of data

* Example:

* User A → sees sales of North region

# Q10.Assume an organization shares reports without using Apps, workspace roles, or RLS. Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain how Power BI best practices help prevent them ?

ANSWER

1. Risks of Sharing Reports Without Apps, Workspace Roles, or RLS in Microsoft Power BI

2. Unauthorized Data Access

* Issue: Without proper controls, users may see sensitive or confidential data that they should not access.

* Example: An employee from one department can view another department’s financial or HR data.

* How Best Practices Prevent This:

Use Row-Level Security (RLS) to restrict data

Assign workspace roles (Viewer, Member, Admin)

Share reports through Apps for controlled access



3. Lack of Governance & Control

*  Issue: No clear ownership or control over reports leads to confusion and mismanagement.

* Problems:

 Multiple versions of the same report

Unauthorized edits

No tracking of changes

* How Best Practices Prevent This:

Use Apps to publish a single official version

Limit editing rights to specific users

Maintain a proper development → testing → production workflow




4. Data Misuse or Oversharing

* Issue: Users may download, export, or forward reports without restrictions.

* Risks:

Data leaks outside the organization

Misinterpretation of incomplete data

Sharing outdated or incorrect reports

* How Best Practices Prevent This:

Apply access permissions and restrictions

Control export/download settings

Use Apps to distribute only approved content
