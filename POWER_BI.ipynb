# Question 1 : What is Power BI and why is it used in businesses?

ANSWER  

1. Power BI is software that

* Collects data from different sources (Excel, databases, websites, etc.)

* Converts raw data into charts, graphs, and dashboards

* Helps users understand data easily

* Example: Sales data can be shown as a colorful chart instead of numbers.

2.  Better decision-making

* Helps managers understand company performance

* They can make smart decisions based on data

* Example: Identify which product sells the most

3.  Easy data visualization

* Converts complex data into simple charts and graphs

* Easy to understand

4.  Combines data from many sources

* Excel

* SQL databases

* Online services

* All data can be seen in one place

# Question 2 : Name and explain the three main components of Power BI

ANSWER

1. Power BI Desktop

* Power BI Desktop is a software application installed on a computer.

* It is used to create reports and dashboards.

* Users can connect, clean, and visualize data here.

* Example: Creating sales charts using Excel data.


2. Power BI Service

* Power BI Service is an online platform (cloud-based).

* It is used to publish, share, and view reports and dashboards.

* It allows teams to access reports from anywhere using the internet.


3. Power BI Mobile

* Power BI Mobile is a mobile app available for phones and tablets.

* It helps users view reports and dashboards on mobile devices.

* Useful for managers to check business data anytime, anywhere.

# Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view

ANSWER

1. Data View

* Data View shows the data in table format (rows and columns).

 Purpose:

* To view and understand the data

* To check values in each column

* To create new columns and calculations

* Example: Viewing property prices in the real estate dataset.


2. Report View

* Report View is used to create charts, graphs, and visual reports.

 Purpose:

* To create visualizations

* To design reports and dashboards

* To analyze data visually

* Example: Creating a bar chart showing price by city.

3. Model View

* Model View shows the relationship between different tables.

Purpose:

* To connect tables

* To manage relationships

* To create proper data model

* Example: Connecting Customer table with Property table
