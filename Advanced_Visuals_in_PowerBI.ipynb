# Q1 What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?
 ANSWER


1. Basic filter
* purpose = Filters data by selecting values from a list
* work = You check/uncheck items in a list
* user interaction = Mostly used by report creators in the filter pane
* complexity = simmple
* example = Show sales only for India

2. advance filter
* purpose =  Filters data using conditions (rules)
* work =  You apply conditions like contains, greater than, less than
* user interaction =  Used by report creators for complex filtering
* complexity =  More powerful and flexible
* example = show sales greater than 10,000 or name contains “A”

3. slicer
* purpose = Visual filter placed on the report page
* work = Users interact with buttons, dropdowns, or lists
* user interaction = used by report viewers directly on the report
* complexity = Simple and interactive
* example = A visual button list where users click Year = 2024

# Q4 What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?
. ANSWER


1. visual level filter
* A Visual-Level Filter applies only to one specific visual chart, table,

* Example: If you apply a filter Region = North to a Pie Chart, only that pie chart will show North region data, while other visuals on the page remain unchanged.


2. page level filter

* A Page-Level Filter applies to all visuals on the same report page.

* Example: If you set Year = 2023 as a page filter, every chart and table on that page will display only 2023 data.

3.. report level filter
* A Report-Level Filter applies to all pages in the entire report.

* Example: If you apply Country = India as a report filter, every page and visual in the report will show only India data.

# Q6 What is the role of Drill-Down in Power BI? How does it improve data exploration?

6. ANSWER
1. drill down =  Drill-Down is a feature that allows users to explore data in more detail by moving from higher-level data to lower-level data in a hierarchy


2. Role of Drill-Down = Drill-Down helps users break down summarized data into detailed levels. For example, you can move from:

* Year → Quarter → Month → Day

* Category → Product

* This allows users to analyze data step-by-step.

3. How it Improves Data Exploration

* Better understanding of data – Users can see detailed information behind summary values.

* Interactive analysis – Users can click on charts to explore deeper levels.

* Find patterns and trends – Helps identify which product, region, or time period contributes to results

# Q13 What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction

. ANSWER

1. the “Keep All Filters” option in Drill-Through controls whether the filters from the original page are carried over to the drill-through page.

2. What “Keep All Filters” Does

* When Keep All Filters is ON, all filters applied on the source page (such as slicers, page filters, or visual filters are passed to the drill-through page along with the selected field.

3. Example

Suppose a report has:

Region slicer = East

Year filter = 2023

A chart showing Products

4. How it Affects User Interaction

* Maintains context – Users continue analyzing data with the same filters applied.

* More precise analysis – Detailed pages reflect the exact filtered view.

* Better user experience – Users don’t need to reapply filters on the drill-through page.
